```python
a=0.01
CS=pd.DataFrame([[10000,0,0,0,0,0] for i in range(40)],columns=['C','shares1','shares2','shares3','shares4','shares5'])
```

```python
#现金变成股票
#现金*(1-a)/股价
def Shares(C,V,a):
    z=C*(1-a)/V
    return z

#股票变成现金
#股票*股价*(1-a)
def Cell(S,V,a):
    z=S*(1-a)*V
    return z
```

```python
C=10000
S1=0
S2=0
S3=0
S4=0
S5=0
temp=0
```

```python
for i in range(58):
    #交易次数
    #当天股价
    S1_shares_today=data1['Open'][i]
    S2_shares_today=data2['Open'][i]
    S3_shares_today=data3['Open'][i]
    S4_shares_today=data4['Open'][i]
    S5_shares_today=data5['Open'][i]
    
    #明天股价
    S1_shares_tomo=data1['Open'][i+1]
    S2_shares_tomo=data2['Open'][i+1]
    S3_shares_tomo=data3['Open'][i+1]
    S4_shares_tomo=data4['Open'][i+1]
    S5_shares_tomo=data5['Open'][i+1]
      
    #卖
    if(S1!=0):
        if(data1['Open'][i]>data1['Open'][i+1]):
            CS['C'][temp+1]=Cell(S1,S1_shares_today,a)+C
            CS['shares1'][temp+1]=0
            temp=temp+1
    if(S2!=0):        
        if(data2['Open'][i]>data2['Open'][i+1]):
            CS['C'][temp+1]=Cell(S2,S2_shares_today,a)+C
            CS['shares2'][temp+1]=0
            temp=temp+1
    if(S3!=0):        
        if(data3['Open'][i]>data3['Open'][i+1]):
            CS['C'][temp+1]=Cell(S3,S3_shares_today,a)+C
            CS['shares3'][temp+1]=0
            temp=temp+1
    if(S4!=0):        
        if(data4['Open'][i]>data4['Open'][i+1]):
            CS['C'][temp+1]=Cell(S4,S4_shares_today,a)+C
            CS['shares4'][temp+1]=0
            temp=temp+1
    if(S5!=0):        
        if(data5['Open'][i]>data5['Open'][i+1]):
            CS['C'][temp+1]=Cell(S5,S5_shares_today,a)+C
            CS['shares5'][temp+1]=0
            temp=temp+1
            
    C=CS['C'][temp]     
            
        #分别买五支股票的数量
    S1_quantity=Shares(C,S1_shares_today,a)
    S2_quantity=Shares(C,S2_shares_today,a)
    S3_quantity=Shares(C,S3_shares_today,a)
    S4_quantity=Shares(C,S4_shares_today,a)
    S5_quantity=Shares(C,S5_shares_today,a)
    
    #预测购买五支股票的价值
    C_pre1=Cell(S1_quantity,S1_shares_tomo,a)
    C_pre2=Cell(S2_quantity,S2_shares_tomo,a)
    C_pre3=Cell(S3_quantity,S3_shares_tomo,a)
    C_pre4=Cell(S4_quantity,S4_shares_tomo,a)
    C_pre5=Cell(S5_quantity,S5_shares_tomo,a)
    
    #分别收益
    profit1=C_pre1-C
    profit2=C_pre2-C
    profit3=C_pre3-C
    profit4=C_pre4-C
    profit5=C_pre5-C
    
    profit=pd.DataFrame([profit1,profit2,profit3,profit4,profit5]).sort_values(by=0)
    profit_pos=profit.loc[profit[0]>=0]
    profit_sum=profit_pos[0].sum()
    #买(比较收益！！！)
    if(C!=0):
        if(profit.iloc[-1,0]>0):
            if(profit1>=0):
                S1_true=Shares(C*profit1/profit_sum,S1_shares_today,a)
                CS['shares1'][temp+1]=S1_true+S1
            if(profit2>=0):
                S2_true=Shares(C*profit2/profit_sum,S1_shares_today,a)
                CS['shares2'][temp+1]=S2_true+S2
            if(profit3>=0):
                S3_true=Shares(C*profit3/profit_sum,S1_shares_today,a)
                CS['shares3'][temp+1]=S3_true+S3
            if(profit4>=0):
                S4_true=Shares(C*profit4/profit_sum,S1_shares_today,a)
                CS['shares4'][temp+1]=S4_true+S4
            if(profit5>=0):
                S5_true=Shares(C*profit5/profit_sum,S1_shares_today,a)
                CS['shares5'][temp+1]=S5_true+S5
            CS['C'][temp+1]=0
            temp=temp+1
            
    #持有现金和股票数量
    C=CS['C'][temp]
    S1=CS['shares1'][temp]
    S2=CS['shares2'][temp]
    S3=CS['shares3'][temp]
    S4=CS['shares4'][temp]
    S5=CS['shares5'][temp]
    
    
#卖
if(S1!=0):
    CS['C'][temp+1]=Cell(S1,S1_shares_today,a)+C
    CS['shares1'][temp+1]=0
    temp=temp+1
if(S2!=0):        
    CS['C'][temp+1]=Cell(S2,S2_shares_today,a)+C
    CS['shares2'][temp+1]=0
    temp=temp+1
if(S3!=0):        
    CS['C'][temp+1]=Cell(S3,S3_shares_today,a)+C
    CS['shares3'][temp+1]=0
    temp=temp+1
if(S4!=0):        
    CS['C'][temp+1]=Cell(S4,S4_shares_today,a)+C
    CS['shares4'][temp+1]=0
    temp=temp+1
if(S5!=0):        
    CS['C'][temp+1]=Cell(S5,S5_shares_today,a)+C
    CS['shares5'][temp+1]=0
    temp=temp+1
```

```python
#调用库
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.preprocessing import MinMaxScaler
from keras import regularizers
```

```python
#读入数据
MKPRU_Value = pd.read_csv(r'BCHAIN-MKPRU.csv')
MKPRU_Value
```

```python
#时间长度
time_stamp = 5 # 输入序列长度
```

```python
for temp in range(51,1800):
    data = MKPRU_Value[['Value']]
    train_i=int(0.8*temp)
    valid_i=temp
    test_i=temp
    # 划分训练集与验证集 
    train = data[0:train_i + time_stamp]   #训练集，修改其中数字
    valid = data[train_i - time_stamp:valid_i + time_stamp]   #验证集
    test = data[test_i- time_stamp:test_i+1+time_stamp]   #测试集
    
    #找测试集中最大值最小值
    a=float(test.min())
    b=float(test.max())
    
    #定义归一化函数
    scaler = MinMaxScaler(feature_range=(0, 1))
    
    # 将训练集归一化
    scaled_data = scaler.fit_transform(train)    #将每个train归一化
    x_train, y_train = [], []
    
    # 训练集切片
    for i in range(time_stamp, len(train)-5):
        x_train.append(scaled_data[i - time_stamp:i])
        y_train.append(scaled_data[i: i+5])
     
    x_train, y_train = np.array(x_train), np.array(y_train).reshape(-1,5)
    
    #验证集归一化
    scaled_data = scaler.fit_transform(valid)
    x_valid, y_valid = [], []
    
    # 验证集切片
    for i in range(time_stamp, len(valid)-5):
        x_valid.append(scaled_data[i - time_stamp:i])
        y_valid.append(scaled_data[i: i+5])

    x_valid, y_valid = np.array(x_valid), np.array(y_valid).reshape(-1,5)
    
    # 测试集归一化
    scaled_data = scaler.fit_transform(test)
    x_test, y_test = [], []
    
    # 测试集切片
    for i in range(time_stamp, len(test)-5):
        x_test.append(scaled_data[i - time_stamp:i])
        y_test.append(scaled_data[i: i+5])

    x_test, y_test = np.array(x_test), np.array(y_test).reshape(-1,5)
    
    #### 建模部分 ####
    #训练集数据构建网络
    model = keras.Sequential()   
    model.add(layers.LSTM(64, return_sequences=True, input_shape=(x_train.shape[1:]))) #第一层网络
    model.add(layers.LSTM(64, return_sequences=True))  #第二层网络
    model.add(layers.LSTM(32))  #第三层网络
    model.add(layers.Dropout(0.1))  #第四层网络，第四层为dropout层(dropout=0.054）前三层为Lstm
    model.add(layers.Dense(5))  #第四层为全连接层(神经元个数为5)

    #损失函数
    model.compile(optimizer=keras.optimizers.Adam(), loss='mae',metrics=['accuracy'])


    learning_rate_reduction = keras.callbacks.ReduceLROnPlateau(monitor='val_loss', patience=3, factor=0.7, min_lr=0.000000005)
    #回调函数ReduceLROnPlateau
    #monitor：监测的值，可以是accuracy，val_loss,val_accuracy
    #factor：缩放学习率的值，学习率将以lr = lr*factor的形式被减少
    #patience：当patience个epoch过去而模型性能不提升时，学习率减少的动作会被触发
    #mode：‘auto’，‘min’，‘max’之一 默认‘auto’就行
    #epsilon：阈值，用来确定是否进入检测值的“平原区”
    #cooldown：学习率减少后，会经过cooldown个epoch才重新进行正常操作
    #min_lr：学习率最小值，能缩小到的下限



    history = model.fit(x_train, y_train,batch_size = 128,epochs=70,
                        validation_data=(x_valid, y_valid),callbacks=[learning_rate_reduction])

    # loss变化趋势可视化
    #plt.plot(history.history['loss'],label='训练损失')   #history对象
    #plt.plot(history.history['val_loss'], label='实际损失')
    #plt.show()
    
    #用训练集训练出模型让损失最小
    
    #### 预测结果分析&可视化 #### 

    closing_price = model.predict(x_test)      #输入样本输出预测
    model.evaluate(x_test)  #输入数据(data)和真实标签(label),然后将预测结果与真实标签相比较,得到两者误差并输出.
    scaler.fit_transform(pd.DataFrame(valid['Value'].values))  #统一处理数据数据

    # 反归一化-得到数据实际的值
    closing_price = scaler.inverse_transform(closing_price.reshape(-1,5)[:,0].reshape(1,-1)) #只取第一列
    y_test = scaler.inverse_transform(y_test.reshape(-1,5)[:,0].reshape(1,-1))

    # 计算预测结果
    #rms = np.sqrt(np.mean(np.power((y_test[0:1,5:] - closing_price[0:1,5:] ), 2)))  #均方根
    #print(rms)                   #计算值
    #print(closing_price.shape)   #预测值
    #print(y_test.shape)          #测试集中的实际值

    # 预测效果可视化
    #plt.figure(figsize=(16, 8))
    dict_data = {
        'Predictions': closing_price.reshape(1,-1)[0],
        'Value': y_test[0]
    }
    data_pd = pd.DataFrame(dict_data)
    #plt.plot(data_pd[['Value']],linewidth=3,alpha=0.8)
    #plt.plot(data_pd[['Predictions']],linewidth=1.2)
    
    #plt.show()
    print(temp)
    scaler = MinMaxScaler(feature_range=(a,b))
    scaled_data = scaler.fit_transform(data_pd)
    scaled_data
    
    scaled_data = pd.DataFrame(scaled_data)
    #plt.figure(figsize=(16, 8))
    #plt.plot(scaled_data[0],linewidth=3,alpha=0.8)
    #plt.plot(scaled_data[1],linewidth=1.2)

    #plt.show()
    for j in range(len(scaled_data[0])):
        scaled_data_ar.append(scaled_data[0][j])
```
